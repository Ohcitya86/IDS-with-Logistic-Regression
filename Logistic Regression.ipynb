```python
correlation_threshold = 0.1
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
```

```python
weak_correlations = (df[numeric_columns].abs() < correlation_threshold) & (df[numeric_columns].abs() > -correlation_threshold)

```

```python
columns_to_keep = df[numeric_columns].columns[~weak_correlations.any()]
```

```python
new_df = df[columns_to_keep]
```

```python
new_df
```

```python
weak_correlations_mask = (df[numeric_columns].abs() < correlation_threshold) & (df[numeric_columns].abs() > -correlation_threshold)
```

```python
weak_correlations_values = df[numeric_columns][weak_correlations_mask]
```

```python
print("Weak Correlations:")
print(weak_correlations_values)
```

```python
strong_correlations_mask = (df[numeric_columns].abs() > correlation_threshold)
```

```python
strong_correlations_values = df[numeric_columns][strong_correlations_mask]

# Display the strong correlations
print("Strong Correlations:")
print(strong_correlations_values)
```
